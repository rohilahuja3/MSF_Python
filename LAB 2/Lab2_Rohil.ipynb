```python
def Tax(amt, rate):
    tax_amt = amt*rate
    return tax_amt

def main():
    taxrate = 0.08  
    purchase = float(input('Enter the purchase amount: '))
    
    #printing the input, my IDE doesn't support displaying input
    print("Enter the purchase amount: {}".format(int(purchase)))
    
    tax=Tax(purchase,taxrate)
    total=tax+purchase
    print ('Purchase amount: ${:,.2f}'.format(purchase))
    print ('Sales tax: ${:,.2f}'.format(tax))
    print('Total purchase: ${:,.2f}'.format(total))

main()
```

```python
def FV(p, r, t):
    total = p*(1+r/100)**t
    return total

def main():
    p = float(input("Enter principal: "))
    r = float(input("Enter interest rate: "))
    t = int(input("Enter # of periods: "))

    #printing the input, my IDE doesn't support displaying input
    print("Enter principal: {}".format(int(p)))
    print("Enter interest rate: {}".format(r))
    print("Enter # of periods: {}".format(t))

    total = FV(p, r, t)
    print("The future value is: ${:,.2f}".format(total))

main()
```

```python
# 0 == heads; 1 == tails

import random as rnd

def headORtail(sim):
    head_count = 0
    for i in range(sim):
        val = rnd.randint(0,1)
        if (val==0):
            head_count+=1
    return head_count

def main():
    sim = int(input("Enter number of coin flips: "))
    
    #printing the input, my IDE doesn't support displaying input
    print("Enter number of coin flips: {}".format(int(sim)))

    x = headORtail(sim)
    print("You had {} heads out of {} coin flip.".format(x, sim))

main()
```

```python
def atm(upper, lower):
    while True:
        amt = float(input("Enter the withdrawal amount: "))

        #printing the input, my IDE doesn't support displaying input
        print("Enter the withdrawal amount: {}".format(int(amt)))

        if (upper-amt < lower):
            print("You have reached or exceeded your floor balance of ${:,.2f}".format(lower))
            break
        else:
            upper-=amt

def main():
    upper = float(input("Enter Account Balance: "))
    lower = float(input("Enter Lower Bound: "))

    #printing the input, my IDE doesn't support displaying input
    print("Enter Account Balance: {}".format(int(upper)))
    print("Enter Lower Bound: {}".format(int(lower)))

    atm(upper, lower)

main()
```

```python
# simple function to calculate the backdated PV of $100
def PV(p, r, t):
    PV = p/(1+r)**t
    return PV

def main():
    p = 100 # Given principal amount
    r = 2.5/100 # Rate

    # Using a infinite while loop to continiously ask for years and if the user wants to continue or not
    while True:
        year = int(input("Enter year of birth: "))
        #printing the input, my IDE doesn't support displaying input
        print("Enter year of birth: {}".format(year))

        # First logic, to check if the year enterned is within the range 
        if (year<1899 or year>2024):
            print("You need to enter a year before 2024 and after 1899.\n")
            continue # if year is not within the range, it jumps back to the while True statment (or restarts the loop), and ignores everything else in the loop

        # calculates the number or periods and calls the present value function
        t = 2024-year
        present_val = PV(p, r, t)
        print("\n${} in 2024 was worth ${:,.2f} in {}\n".format(p, present_val, year))

        # User input validation for continuing 
        resp = input("Do you want to enter another year? y for yes: ").lower()
        #printing the input, my IDE doesn't support displaying input
        print("Do you want to enter another year? y for yes: {}\n".format(resp))

        # For an infinite number of user inputs, this is a validating step to ask the user if they want to continue or not
        if (resp!='yes' and resp!='y'):
            # if the user inputs anything other than yes, the chat or loop ends.
            print("Ending the chat!")
            break

main()
```

```python
def FV_fun(PV, arate, PMT):
    r = arate/1200 # To convert the annual rate to monthly rate
    FV = PV # assigning the FV as PV and we will calculate the FV of 10,000 after annuity payout and including the interest
    t=0 # number of months counter

    while (FV>PMT): # check if the annuity payment is less than initial investment
        t+=1
        FV_of_PV_t = PV*(1+r)**t # this calculates the FV after the every month of compounding
        FV_of_PMT_t = PMT*(((1+r)**t) -1)/r # this calculates the PMT compounding after every month
        FV = FV_of_PV_t-FV_of_PMT_t # this calculates the difference between FV and PMT of each month, at the end it stores the last month FV (or balance of the initial sum)
    print("After {} months your balance is ${:,.2f}.".format(t, FV))

def main():
    pv = float(input("Please enter your initial investment: "))
    arate = float(input("Please enter annual interest rate (ex. 2.5%): "))
    pmt = float(input("Please enter the monthly annuity payout: "))
    
    #printing the input, my IDE doesn't support displaying input
    print("Please enter your initial investment: ${:,.2f}\n".format(pv))
    print("Please enter annual interest rate (ex. 2.5%): {}%\n".format(arate))
    print("Please enter the monthly annuity payout: ${:,.2f}".format(pmt))
    
    FV_fun(pv, arate, pmt)

main()
```

```python
def depreciation(item_name, cost, life, method):
    
    print("Depreciation schedule for: {}\n".format(item_name))
    print("Year"+"\t\tBegin"+"\t\tDep"+"\t\tEnd")

    book_val = cost # book_val = cost = 1500

    if method==1: # True
        simple_rate = 1/life # 20% on the cost of the item, constant each year
    else:
        double_rate = 2/life # 40% on the ending year value of the previous year, varies every year

    for year in range(life+1): # year = 0,1,2,3,4,5 -> total 6 iterations, including the year 0, when dep is 0 irrespective of method
        if method==1:
            if year==0:
                dep = 0          # year = 0, dep = 0
            else:
                dep = simple_rate*cost # Because the depreciation in method 1 remians the same, therefore multiply by cost = 300
        else: # method == 2 
            # We have to consider 3 conditions:
                # 1) In the final year of life, the depreciation is the ending value in the previous year
            if year==life:
                dep = book_val
                # 2) To print the base condition when year = 0, dep = 0
            elif year==0:
                dep = 0 
                # 3) Not only the depreciation is higher, its calculated on the ending book value of the previous year. It is not a constant value like method 1
            else:
                dep = double_rate*book_val

        end_val = book_val-dep # Calculating the ending value of the year, book_val (year starting val) - dep
        print("{}\t\t{:.2f}\t\t{:.2f}\t\t{:.2f}".format(year, book_val, dep, end_val))
        # Now we have to assign the next year beg book val as the previous year ending value. Irrespective of the method.
        book_val = end_val
    
def main():
    item = input("Please enter a description of the item: ")
    cost = float(input("Please enter the cost of the item: "))
    life = int(input("Please enter the estimated life of the item in whole years: "))
    method = int(input("Please enter depreciation method (1 = straight line, 2 = double declining balance): "))

    #printing the input, my IDE doesn't support displaying input
    print("Please enter a description of the item: {}\n".format(item))
    print("Please enter the cost of the item: ${:,.2f}\n".format(cost))
    print("Please enter the estimated life of the item in whole years: {}\n".format(life))
    print("Please enter depreciation method (1 = straight line, 2 = double declining balance): {}".format(method))

    depreciation(item, cost, life, method)

main()
```
