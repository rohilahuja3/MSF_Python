```python
def main():
    try:
        total = int(input("Enter total: "))
        print("Enter total: {}".format(total))
        num_item = int(input("Enter num: "))
        print("Enter num: {}".format(num_item))
        avg = total/num_item
        print(avg)
    except ZeroDivisionError:
        print("ERROR: cannot have 0 items")
    except ValueError:
        print("ERROR: number of items cannot be negavtive")
main()
```
