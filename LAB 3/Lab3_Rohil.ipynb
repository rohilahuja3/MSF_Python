```python
def fun1(user_input):
    # Check if string is alphanumeric -> contains only letters and/or numbers, no spaces, symbols, or punctuation
    if user_input.isalnum():
        print("Is alphanumeric")

    # Check if string is only digits
    if user_input.isdigit():
        print("Contains only numbers")

    # Check if string is only alphabetic
    if user_input.isalpha():
        print("Contains only letters")
        
        # if only alphabetic then check if the string is all uppercase
        if user_input.isupper():
            print("Is in all CAPS")

        # if only alphabetic then check if the if string is all lowercase
        elif user_input.islower():
            print("Is all lowercase")

def main():
    user_input = input("Enter your string: ")
    #printing the input, my IDE doesn't support displaying input
    print("Enter your string: {}".format(user_input))
    print("Your string:")
    fun1(user_input)

main()
```

```python
def fun_strip(amt):
    amt_str = amt.strip('$')
    half_amt = float(amt_str)/2

    print("\nStripping Method")
    print("Half of that amount is: ${:,.2f}".format(half_amt))

def fun_slice(amt):
    amt_str = amt[1:]
    half_amt = float(amt_str)/2

    print("\nSlicing Method")
    print("Half of that amount is: ${:,.2f}".format(half_amt))


def main():
    amt = input("Enter the assest value in the format $XXXX: ")
    #printing the input, my IDE doesn't support displaying input
    print("Enter the assest value in the format $XXXX: {}".format(amt))

    fun_strip(amt)
    fun_slice(amt)

main()
```

```python
stock_num = int(input("Enter number of stocks in portfolio: "))
#printing the input, my IDE doesn't support displaying input
print("Enter number of stocks in portfolio: {}".format(stock_num))

# Initialize lists to store stock prices, share number, and per-stock total values
stock_prices = []
stock_numbers = []
total_list = []

# Collect the price for each stock
for i in range(stock_num):
    stock_price = float(input('Enter the price of stock {}: '.format(i+1)))
    #printing the input, my IDE doesn't support displaying input
    print("Enter the price of stock {}: \n{}".format(i+1, int(stock_price)))
    stock_prices.append(stock_price) # append in the empty list

# Collect the number of shares for each stock
for i in range(stock_num):
    stock_number = float(input('Enter the # shares of stock {}: '.format(i+1)))
    #printing the input, my IDE doesn't support displaying input
    print("Enter the # shares of stock {}: \n{}".format(i+1, int(stock_number)))
    stock_numbers.append(stock_number) # append in the empty list

# Calculate and print the holding value for each stock
for i in range(len(stock_prices)):
    total = stock_prices[i]*stock_numbers[i]
    total_list.append(total) # append in the empty list
    print("Your holding for stock {} = ${}".format(i+1, int(total_list[i]))) 

# Calculate and print the total value of the entire portfolio
x = sum(total_list)
print("The total value of your portfolio is: ${}".format(int(x)))
```

```python
stock_num = int(input("Enter number of stocks in portfolio: "))
#printing the input, my IDE doesn't support displaying input
print("Enter number of stocks in portfolio: {}\n".format(stock_num))

# Initialize lists to store stock names, prices, share counts, and total values
stock_names = []
stock_prices = []
stock_numbers = []
total_list = []

# Collect names of each stock
for i in range(stock_num):
    stock_name = input('Enter the name of stock {}: '.format(i+1))
    #printing the input, my IDE doesn't support displaying input
    print("Enter the name of stock {}: \n\n{}\n".format(i+1, stock_name))
    stock_names.append(stock_name) # append in the empty list

# Collect prices for each stock
for i in range(stock_num):
    stock_price = float(input('Enter the price of {}: '.format(stock_names[i])))
    #printing the input, my IDE doesn't support displaying input
    print("Enter the price of stock {}: \n\n{}\n".format(i+1, int(stock_price)))
    stock_prices.append(stock_price) # append in the empty list

# Collect number of shares for each stock
for i in range(stock_num):
    stock_number = float(input('Enter the # shares of {}: '.format(stock_names[i])))
    #printing the input, my IDE doesn't support displaying input
    print("Enter the # shares of stock {}: \n\n{}\n".format(i+1, int(stock_number)))
    stock_numbers.append(stock_number) # append in the empty list

# Calculate the total value per stock and print individual holdings
for i in range(len(stock_prices)):
    total = stock_prices[i]*stock_numbers[i]
    total_list.append(total) # append in the empty list
    print("Your holding for stock {} = ${:,.0f}\n".format(stock_names[i], int(total_list[i]))) 

# Calculate and print the total value of the entire portfolio
x = sum(total_list)
print("The total value of your portfolio is: ${:,.0f}".format(int(x)))
```

```python
# Import the pandas library
import pandas as pd

# Load the dataset from the CSV file into a DataFrame
df = pd.read_csv("loan default.csv")

# Filter the data to include only buyers over the age of 50
df_new = df[df['Bo_Age']>50]

# Group the filtered data by loan OUTCOME (default / non-default) and count the number of buyers (using Cust_ID)
count_by_outcome = df_new.groupby('OUTCOME')['Cust_ID'].count()
print(count_by_outcome,'\n\n\n')

# Calculate the average (mean) of all numeric columns for each OUTCOME group in the filtered df_new
mean_by_outcome = df_new.groupby('OUTCOME').mean(numeric_only=True)
print(mean_by_outcome)
```
