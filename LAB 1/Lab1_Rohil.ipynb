```python
n = int(input("Enter the number of assets: "))
#printing the input, my IDE doesn't support displaying input
print("Enter the number of assets: {}".format(n))
total = 0 # declaring var as 0

for i in range(n): # for loop runs
    val = float(input("Enter asset {} value: ".format(i+1))) # i+1 because i starts at 0 index, intuitive for user
    #printing the input, my IDE doesn't support displaying input
    print("Enter asset {} value: {}".format(i+1, val))
    total += val
    
print("The total value of the {} assets is: ${}".format(n,total))
```

```python
mile = float(input("Enter miles: "))
yard = float(input("Enter yards: "))
ft = float(input("Enter feet: "))
inch = float(input("Enter inches: "))

#printing the input, my IDE doesn't support displaying input
print("Enter miles: {}".format(mile))
print("Enter yards: {}".format(yard))
print("Enter feet: {}".format(ft))
print("Enter inches: {}".format(inch))

cm2inch = 0.3937 # 1 cm = 0.3937 inch
inch2cm = 1/cm2inch # 1 inch = 2.54 cm
ft2cm = 12*inch2cm # 1 feet = 12 inch = 30.48 cm
yard2ft = 3*ft2cm # 1 yard = 3 feet = 91.44 cm
mile2yard = 1760*yard2ft # 1 mile = 1760 yard = 160,934.4 cm
m2cm = 100 # 1 m = 100 cm
km2m = 1000*m2cm # 1 km = 1000 m = 1000*100 cm
cm2m = 1/m2cm # 1 cm = 0.01 m
cm2km = 1/km2m # 1 cm = 0.00001 km

total_cm = mile * mile2yard + yard * yard2ft + ft * ft2cm + inch * inch2cm # converting all into cm

km = int(total_cm // km2m) # cm to km conversion
m = int((total_cm % km2m) // m2cm) # cm to m conversion
cm = round(total_cm % m2cm, 1)

print(format(km),"kilometers")
print(format(m),"meters")
print(format(cm),"centimeters")
```

```python
rate = float(input("Enter the exchange rate (1 USD = ? EUR): "))
curr = input("Do you wish to exchange dollars or euros? ").lower()
fee = 6.9/100

#printing the input, my IDE doesn't support displaying input
print("Enter the exchange rate (1 USD = ? EUR): {}".format(rate))
print("Do you wish to exchange dollars or euros? {}".format(curr))

if curr == "usd" or curr == "dollar": # extreme case handling
    amt = float(input("Please enter the number of US Dollars to exchange: "))
    print("Please enter the number of US Dollars to exchange: ${}".format(amt))

    total = amt*rate # converts the total converted val
    transaction_fee = total*fee # cal the fee
    after_fee_total = total - transaction_fee # sub the free from total amount
    
    print("The total exchange value is {}".format(round(total, 2)))
    print("The transaction fee is {}".format(round(transaction_fee, 2)))
    print("You will receive a total of {}".format(round(after_fee_total, 2)))

elif curr == "eur" or curr == "euros" or curr == "euro": # extreme case handling
    amt = float(input("Please enter the number of Euros to exchange: "))
    print("Please enter the number of Euros to exchange: {}".format(amt))

    total = amt/rate # converts the total converted val
    transaction_fee = total*fee # cal the fee
    after_fee_total = total - transaction_fee # sub the free from total amount

    print("The total exchange value is ${}".format(round(total, 2)))
    print("The transaction fee is ${}".format(round(transaction_fee, 2)))
    print("You will receive a total of ${}".format(round(after_fee_total, 2)))

else:
    print("This is not a valid selection.")
```
